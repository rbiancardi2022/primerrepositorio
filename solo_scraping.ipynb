```python
lista_tkt = ['DDD', 'AYX', 'SQ', 'CDNA', 'COIN', 'COMM', 'DDOG', 'DM', 'DKNG', 'DOCU', 'ESTC', 'FIVN', 'HCAT', 'INTC', 'KTOS', 'MRVL',
             'MGM', 'MDB', 'PANW', 'RVNC', 'RBLX', 'ROKU', 'SHOP', 'SNAP', 'SNOW', 'SPLK', 'SPOT', 'STEM', 'TTWO', 'TDOC', 'BA', 'GPS',
            'TWLO', 'UBER', 'PATH', 'U', 'WBA', 'WDAY', 'ZM']  

for tkt in lista_tkt:
    try:
        web = f'https://finance.yahoo.com/quote/{tkt}/financials?p={tkt}'
        path = 'C:/Users/VivoBook/Documents/Roberto/__web_scraping/chromedriver.exe'
        service = Service(path)
        driver = webdriver.Chrome(service=service)
        driver.get(web)
        driver.maximize_window
        time.sleep(7)

    #    cerrar_banner = driver.find_element(By.XPATH, '//button[contains(@class,"Pos(a) T(20px) End(20px) P(4px) Bd(0) Bgc(t) M(0)")]')
    #    cerrar_banner.click()
    #    time.sleep(7)


        quarterly = driver.find_element(By.XPATH, '//button[contains(@class,"P(0px) M(0px) C($linkColor) Bd(0px) O(n)")]')
        quarterly.click()
        time.sleep(7)


        detalle = driver.find_elements(By.XPATH, '//button[contains(@class,"P(0) M(0) Va(m) Bd(0) Fz(s) Mend(2px) tgglBtn")]')
        for ver_detalle in detalle:
            ver_detalle.click()
        time.sleep(7)

        fecha_localiza = driver.find_elements(By.XPATH, '//div[contains(@class,"Ta(c) Py(6px) Bxz(bb) BdB Bdc($seperatorColor) Miw(120px) Miw(100px)--pnclg D(ib) Fw(b)")]')
        fecha = fecha_localiza[1].text
        fecha = pd.to_datetime(fecha)
        fecha = fecha.strftime('%Y-%m-%d')

        # valores de las columnas blancas - el primero de cada concepto es el nuevo
        #cajas_1 = driver.find_elements(By.XPATH, '//div[contains(@class,"Ta(c) Py(6px) Bxz(bb) BdB Bdc($seperatorColor) Miw(120px) Miw(100px)--pnclg D(tbc)")]')
        cajas_1 = driver.find_elements(By.XPATH, '//div[contains(@data-test,"fin-col")]')

        cajas_1_valores = []
        for valores_1 in cajas_1:
            cajas_1_valores.append(valores_1.text)

        # los conceptos globales
        cajas_2 = driver.find_elements(By.XPATH,'//div[contains(@class,"D(tbc) Ta(start) Pend(15px)--mv2 Pend(10px) Bxz(bb) Py(8px) Bdends(s) Bdbs(s) Bdstarts(s) Bdstartw(1px) Bdbw(1px) Bdendw(1px) Bdc($seperatorColor) Pos(st) Start(0) Bgc($lv2BgColor) fi-row:h_Bgc($hoverBgColor) Pstart(15px)--mv2 Pstart(10px)")]')
        cajas_2_textos = []
        for textos_2 in cajas_2:
            cajas_2_textos.append(textos_2.text)

        # los conceptos detalles
        cajas_3 = driver.find_elements(By.XPATH, '//div[contains(@class,"D(tbc) Ta(start) Pend(15px)--mv2 Pend(10px) Bxz(bb) Py(8px) Bdends(s) Bdbs(s) Bdstarts(s) Bdstartw(1px) Bdbw(1px) Bdendw(1px) Bdc($seperatorColor) Pos(st) Start(0) Bgc($lv2BgColor) fi-row:h_Bgc($hoverBgColor) Pstart(30px)--mv2 Pstart(25px)")]')
        cajas_3_textos = []
        for textos_3 in cajas_3:
            cajas_3_textos.append(textos_3.text)

        driver.quit()

        cajas_1_valores_final = []
        for i in range(1,len(cajas_1_valores), len(fecha_localiza)): 
            cajas_1_valores_final.append(cajas_1_valores[i])
        df_cajas_1_valores_final = pd.DataFrame({'valor': cajas_1_valores_final})

        cajas_4_textos = cajas_2_textos + cajas_3_textos
        df_cajas_4_textos = pd.DataFrame({'concepto': cajas_4_textos})

        df_cajas_5_textos = pd.merge(df_text_orden, df_cajas_4_textos, on='concepto', how='inner')
        df_cajas_5_textos.reset_index()

        df_tkt = pd.merge(df_cajas_5_textos, df_cajas_1_valores_final, left_index=True, right_index=True )

        mask = df_tkt.concepto.isin(lista_textos_orden_final)

        df_tkt = df_tkt[mask]
        df_tkt = df_tkt.set_index('concepto')
        df_tkt = df_tkt.transpose()
        df_tkt['tkt'] = tkt
        df_tkt['fecha'] = fecha
        df_tkt_parcial = pd.concat([df_tkt_parcial, df_tkt])
    except:
    #    driver.quit()
        pass
```
